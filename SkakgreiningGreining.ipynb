```python
# Skoðum nákvæmni sem fæst með því að nota bara eval mat úr Stockfish

leikir_eval_lina = np.loadtxt("leikir_eval_lina.csv", delimiter=",", dtype=int, comments=None)

leikir_eval_sigurvegari = np.zeros(len(leikir_eval_lina))

for i in range(len(leikir_eval_lina)):
    if leikir_eval_lina[i] > 0:
        leikir_eval_sigurvegari[i] = 1
    elif leikir_eval_lina[i] < 0:
        leikir_eval_sigurvegari[i] = 0
    else:
        leikir_eval_sigurvegari[i] = 2


from sklearn.metrics import accuracy_score

print(accuracy_score(leikir_sigurvegari, leikir_eval_sigurvegari))
```
